```python
import warnings
import pandas as pd
import glob

warnings.filterwarnings("ignore")


csv_files = glob.glob('Final Dataset/*.csv')
row_counts = {}

# Read and concatenate all CSV files
df_list = []
for file in csv_files:
    df = pd.read_csv(file)
    row_counts[file] = len(df)
    df_list.append(df)

combined_df = pd.concat(df_list, ignore_index=True)

# Select the specific columns
selected_columns = ['datetime', 'main.temp', 'main.pressure', 'main.humidity', 'clouds.all', 'weather.main', 'wind.speed', 'city_name']
df = combined_df[selected_columns]

# Display the row counts
print("Row counts per CSV file:")
for file, count in row_counts.items():
    print(f"{file}: {count} rows")

# Display the final dataframe
df
```

```python
 #Data on specific city for example CAMANAVA

cities = ['Caloocan City', 'Malabon', 'Navotas', 'Valenzuela']

# Filter the DataFrame
df_camanava = df[df['city_name'].isin(cities)]

# Display the DataFrame
df_camanava
```

```python
#Checking if there is null value in the dataset
df_camanava.isnull().any()
```

```python
#Count each city
city_counts = df_camanava['city_name'].value_counts()
city_counts
```

```python
df_camanava['weather.main'] = df_camanava['weather.main'].astype(str)

# Extract the 'weather.main' column as a pandas Series
weather_main_series = df_camanava['weather.main']

# Convert the Series to a Python list
weather_main_list = weather_main_series.tolist()

# Print the list
print(weather_main_list)
```

```python
import pandas as pd # type: ignore
# Setting datetime as index
# Convert the 'datetime' column to datetime format
df_camanava['datetime'] = pd.to_datetime(df_camanava['datetime'])
df_camanava.set_index('datetime', inplace=True)
df_camanava.index = df_camanava.index.tz_localize(None)


df_camanava
```

```python
# Weather Counts for validation
clouds_count = weather_main_list.count('Clouds')
rain_count = weather_main_list.count('Rain')
clear_count = weather_main_list.count('Clear')
mist_count = weather_main_list.count('Mist')
smoke_count = weather_main_list.count('Smoke')


print("Clouds count:", clouds_count)
print("Rain count:", rain_count)
print("Clear count:", clear_count)
print("Mist count:", mist_count)
print("Smoke count:", smoke_count)
```

```python
#Using one hot encoder for categorical data
from sklearn.preprocessing import OneHotEncoder# type: ignore

ohe = OneHotEncoder(sparse=False)
one_hot = ohe.fit_transform(df_camanava[['weather.main']])

# Creating new columns for dataframe in weather (Eg. weather.Clear, weather.Clouds, weather.Rain)
columns = ['weather.' + str(category) for category in ohe.categories_[0]]
df_one_hot = pd.DataFrame(one_hot, columns=columns, index=df_camanava.index)

# Replacing the weather.main to one hot encoded 
df_camanava = pd.concat([df_camanava, df_one_hot], axis=1)
df_camanava = df_camanava.drop(['weather.main'], axis=1)

cols = list(df_camanava.columns)
cols.remove('city_name')
df_camanava = df_camanava[cols + ['city_name']]
df_camanava = df_camanava.drop('city_name', axis=1)

df_camanava
```

```python
# Saving the data as a CSV file
import os
file_path = "E:/4-4/regression/final cleaned dataset.csv"
df_camanava.to_csv(file_path, index=True)

```
